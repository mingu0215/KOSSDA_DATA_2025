# 이직원인 + TF-IDF + 이직확률

```python
import pandas as pd

# CSV 불러오기
cause = pd.read_csv("/content/drive/MyDrive/25-1/공모전/posts_tag_이직원인.csv")
prob = pd.read_csv("/content/drive/MyDrive/25-1/공모전/posts_prob.csv")

# 조인 (post id 기준)
merged = cause.merge(prob[["post_id", "이직확률"]], on="post_id", how="left")

# 결과 확인
print(merged.head())

# 저장
merged.to_csv("/content/drive/MyDrive/25-1/공모전/final_text.csv", index=False)
```
