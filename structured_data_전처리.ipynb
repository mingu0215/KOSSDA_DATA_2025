```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 범주형 변수만 추출
category = filled_df.select_dtypes(include=['int']).columns.tolist()

# 유니크 1 이하 제거
cat_cols = [col for col in category if filled_df[col].nunique() > 1]

# Cramér's V 계산 함수 정의
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix, correction=False)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    return np.sqrt(phi2 / min((k - 1), (r - 1)))

# Cramér's V 상관관계 행렬 생성
cramer_matrix = pd.DataFrame(index=cat_cols, columns=cat_cols)

for col1 in cat_cols:
    for col2 in cat_cols:
        if col1 == col2:
            cramer_matrix.loc[col1, col2] = 1.0
        else:
            try:
                cramer_matrix.loc[col1, col2] = cramers_v(filled_df[col1], filled_df[col2])
            except:
                cramer_matrix.loc[col1, col2] = np.nan

plt.figure(figsize=(10, 8))
sns.heatmap(cramer_matrix.astype(float), annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title("Cramér's V Heatmap (Categorical-Categorical Correlation)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
