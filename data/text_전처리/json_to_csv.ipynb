```python
import json
import pandas as pd


jason_path = '/content/drive/MyDrive/25-1/공모전/jobkorea_posts_page_merged.json'

# JSON 파일 불러오기
with open(jason_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 게시글 및 댓글 데이터 추출
posts = []
comments = []

for i, post in enumerate(data, 1):
    post_id = i
    posts.append({
        "post_id": post_id,
        "제목": post["제목"].strip(),
        "내용": post["내용"],
        "태그": post["태그"],
        "작성일": post["작성일"],
        "해시태그": post["해시태그"]
    })

    for comment in post.get("답변들", []):
        comments.append({
            "comment_id": len(comments) + 1,
            "post_id": post_id,  # foreign key
            "작성자": comment.get("작성자", ""),
            "직무/경력": comment.get("직무/경력", ""),
            "레벨": comment.get("레벨", ""),
            "답변": comment.get("답변", ""),
            "작성일": comment.get("작성일", ""),
            "댓글수": comment.get("댓글수", ""),
            "좋아요수": comment.get("좋아요수", "")
        })

# DataFrame 생성
posts_df = pd.DataFrame(posts)
comments_df = pd.DataFrame(comments)

# CSV로 저장
posts_df.to_csv('posts.csv', index=False, encoding='utf-8-sig')
comments_df.to_csv('comments.csv', index=False, encoding='utf-8-sig')

print("✅ posts.csv 와 comments.csv 파일이 생성되었습니다.")
```
